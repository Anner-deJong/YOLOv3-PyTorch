```python
import cv2
from Yolov3 import *
from matplotlib import pyplot as plt
%matplotlib inline
    
yolo    = YOLOv3()
img     = cv2.imread('dog-cycle-car.png') # image from Ayoosh Kathuria
ann_img = yolo.inference(img)

plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(ann_img[:, :, ::-1])
plt.show()
```
